```python
drinks[drinks['country'] == 'South Korea']
```

```python
# etc1 맥주많이먹는 나라들이다.
etc1 = drinks.sort_values(by='beer_servings', ascending=False)
etc1.head(10)
```

```python
# etc2 증류주를 많이먹는 나라들이다.
etc2 = drinks.sort_values(by='spirit_servings', ascending=False)
etc2.head(10)
```

```python
# etc3 와인을 많이먹는 나라들이다.
etc3 = drinks.sort_values(by='wine_servings', ascending=False)
etc3.head(10)
```

```python
# etc4 총알콜량으로 알콜인덱스를 추가해봤다.
# 한국은 27번째로 술을 잘 마시는구나...
q1['Alcohol_index'] = range(len(q1))
q1[q1['country'] == 'South Korea']
```

```python
# etc5 그러면 맥주 인덱스를 만들어보자
# 맥주는 65번째로 잘마시네...
etc1['Beer_index'] = range(len(etc1))
etc1[etc1['country'] == 'South Korea']
```

```python
# etc6 순서대로 증류주와 와인 순서를 출력해보자
# 소주는 129번째로 잘마시고
etc2['Soju_index'] = range(len(etc2))
etc2[etc2['country'] == 'South Korea']
```

```python
# etc7 와인은 91번째네
etc3['Wine_index'] = range(len(etc3))
etc3[etc3['country'] == 'South Korea']
```

```python
# etc8 그러면 한눈에 보고 싶은데...
# 각 술의 전세계 순위 계산
K = q1[q1['country'] == 'South Korea']['Alcohol_index']
L = etc1[etc1['country'] == 'South Korea']['Beer_index']
M = etc2[etc2['country'] == 'South Korea']['Soju_index']
N = etc3[etc3['country'] == 'South Korea']['Wine_index']

# 데이터 정리
korea_rank_values = [K.item(), L.item(), M.item(), N.item()]
labels = ['Alcohol Rank', 'Beer Rank', 'Soju Rank', 'Wine Rank']
```

```python
# 꺾은선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(labels, korea_rank_values, marker='o', color='r', linestyle='-', linewidth=2)
for i, v in enumerate(korea_rank_values):
    plt.text(i, v + 5, str(int(v)), ha='center', va='bottom', fontsize=10)
plt.xlabel('주류별')
plt.ylabel('순위')
plt.title('전세계적으로 봤을 때 한국의 술먹기능력')
plt.gca().invert_yaxis()  # 순위이기 때문에 낮은 숫자가 위에 오도록 y축을 뒤집음
plt.grid(True, axis='y', linestyle='--', linewidth=0.7, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# etc9 그러면 맥주와 소주는 관련있을까?

# 상관 계수 계산
cor1 = drinks['beer_servings'].corr(drinks['spirit_servings'])
cor1
# 중간정도
```

```python
# 대륙별 beer_servings의 평균, 최소, 최대, 합계 시각화
result = drinks.groupby('continent').beer_servings.agg(['mean','min','max','sum'])
result
```

```python
means = result['mean'].tolist()
mins = result['min'].tolist()
maxs = result['max'].tolist()
sums = result['sum'].tolist()

n_groups = len(result.index)

index = np.arange(n_groups)
bar_width = 0.1
graph1 = plt.bar(index, means, bar_width, color='r', label='Mean')
graph2 = plt.bar(index+bar_width, mins, bar_width, color='g', label='Min')
graph2 = plt.bar(index+bar_width * 2, maxs, bar_width, color='b', label='Max')
graph2 = plt.bar(index+bar_width * 3, sums, bar_width, color='y', label='sums')

plt.xticks(index, result.index.tolist())
plt.legend()
plt.show()
```

```python
# 대륙별 알코올 소비량을 시각화하시오, 단, 마지막에 평균 알코올 소비량을 추가할 것
## 전체 알코올 평균
alcohol_mean = drinks.total_litres_of_pure_alcohol.mean()
## 대륙별 알코올 평균
continent_mean = drinks.groupby('continent').total_litres_of_pure_alcohol.mean()
print(continent_mean)
```

```python
continents = continent_mean.index.tolist()
continents.append('mean')
x_pos = np.arange(len(continents))
alcohol = continent_mean.tolist()
alcohol.append(alcohol_mean)
bar_list = plt.bar(x_pos, alcohol, align='center', alpha=0.5)
bar_list[len(continents)-1].set_color('r')
plt.xticks(x_pos, continents)
plt.ylabel('알코올 소비량 평균')
plt.title('대륙별 알코올 소비량')

plt.plot([0,6], [alcohol_mean,alcohol_mean], 'b--')
```
